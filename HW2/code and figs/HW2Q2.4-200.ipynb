```python
#2.4 decision tree classifier with minimum leaf size i criterion.
#this is one typical code for one situation. jusr change the path and get other datasets.
#get from -200.csv

import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score

#this is the path of dataset
Xtrn200 = pd.read_csv('/Users/Agraynel/Desktop/ECE4950/homework/hw2/dataset/X-trn-200.csv', header = None)
Xtst200 = pd.read_csv('/Users/Agraynel/Desktop/ECE4950/homework/hw2/dataset/X-tst-200.csv', header = None)
Ytrn200 = pd.read_csv('/Users/Agraynel/Desktop/ECE4950/homework/hw2/dataset/Y-trn-200.csv', header = None)
Ytst200 = pd.read_csv('/Users/Agraynel/Desktop/ECE4950/homework/hw2/dataset/Y-tst-200.csv', header = None)

entropy_200_train = []
entropy_200_test = []
#change 200, 200 and 400, then we can acquire all the figures of 3 conditions.

for i in range(2, 16):
    clf_entropy = DecisionTreeClassifier(criterion = 'entropy', min_samples_leaf = i, random_state = 0)
    clf_entropy = clf_entropy.fit(Xtrn200, Ytrn200)
    entropy_test = clf_entropy.predict(Xtst200)
    entropy_200_train.append(clf_entropy.score(Xtrn200, Ytrn200))
    entropy_200_test.append(accuracy_score(Ytst200, entropy_test))

# Plotting decision regions

plt.figure(figsize=(15, 5))
plt.plot(range(2, 16), entropy_200_train, c='blue', label='Entropy_train')
plt.plot(range(2, 16), entropy_200_test, c='red', label='Entropy_test')
plt.legend(loc = 3)
plt.ylim(0.75, 1.01)
plt.ylabel('Mean accuracy')
plt.xlabel('Minimum leaf size')
plt.title('2.4 Entropy accuracy: from -200.csv')

plt.show()
```
