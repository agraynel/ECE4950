```python
#2.3 read 50 files decision tree classifier with maximum depth

import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score

#this is the path of dataset
Xtrn50 = pd.read_csv('/Users/Agraynel/Desktop/ECE4950/homework/hw2/dataset/X-trn-50.csv', header = None)
Xtst50 = pd.read_csv('/Users/Agraynel/Desktop/ECE4950/homework/hw2/dataset/X-tst-50.csv', header = None)
Ytrn50 = pd.read_csv('/Users/Agraynel/Desktop/ECE4950/homework/hw2/dataset/Y-trn-50.csv', header = None)
Ytst50 = pd.read_csv('/Users/Agraynel/Desktop/ECE4950/homework/hw2/dataset/Y-tst-50.csv', header = None)

entropy_50_train = []
entropy_50_test = []

for i in range(2, 16):
    clf_entropy = DecisionTreeClassifier(criterion = 'entropy', max_depth= i, random_state = 0)
    clf_entropy = clf_entropy.fit(Xtrn50, Ytrn50)
    entropy_test = clf_entropy.predict(Xtst50)
    entropy_50_train.append(clf_entropy.score(Xtrn50, Ytrn50))
    entropy_50_test.append(accuracy_score(Ytst50, entropy_test))

# Plotting decision regions

plt.figure(figsize=(15, 5))
plt.plot(range(2, 16), entropy_50_train, c='blue', label='Entropy_train')
plt.plot(range(2, 16), entropy_50_test, c='red', label='Entropy_test')
plt.legend(loc=4)
plt.ylim(0.7, 1.01)
plt.ylabel('Mean accuracy')
plt.xlabel('Max tree depth')
plt.title('2.3 Entropy acurracy')

plt.show()
```
