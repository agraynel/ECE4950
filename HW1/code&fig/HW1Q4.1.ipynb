```python
#4.2 read 400 files

import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score

Xtrn400 = pd.read_csv('X-trn-400.csv', header = None)
Xtst400 = pd.read_csv('X-tst-400.csv', header = None)
Ytrn400 = pd.read_csv('Y-trn-400.csv', header = None)
Ytst400 = pd.read_csv('Y-tst-400.csv', header = None)

entropy_400_train = []
entropy_400_test = []
gini_400_train = []
gini_400_test = []

for i in range(2, 16):
    clf_entropy = DecisionTreeClassifier(criterion = 'entropy', max_depth= i, random_state = 0)
    clf_entropy = clf_entropy.fit(Xtrn400, Ytrn400)
    entropy_test = clf_entropy.predict(Xtst400)
    entropy_400_train.append(clf_entropy.score(Xtrn400, Ytrn400))
    entropy_400_test.append(accuracy_score(Ytst400, entropy_test))
    
    clf_gini = DecisionTreeClassifier(criterion = 'gini', max_depth= i, random_state = 0)
    clf_gini = clf_gini.fit(Xtrn400, Ytrn400)
    gini_test = clf_gini.predict(Xtst400)
    gini_400_train.append(clf_gini.score(Xtrn400, Ytrn400))
    gini_400_test.append(accuracy_score(Ytst400, gini_test))

# Plotting decision regions

plt.figure(figsize=(15, 5))
plt.subplot(121)
plt.plot(range(2, 16), entropy_400_train, c='blue', label='Entropy_train')
plt.plot(range(2, 16), entropy_400_test, c='red', label='Entropy_test')
plt.legend(loc=4)
plt.ylim(0.86, 1.01)
plt.ylabel('Mean accuracy')
plt.xlabel('Max tree depth')
plt.title('4.1(b) Entropy acurracy')

plt.subplot(122)
plt.plot(range(2, 16), gini_400_train, c='blue', label='Gini_train')
plt.plot(range(2, 16), gini_400_test, c='red', label='Gini_test')
plt.legend(loc=4)
plt.ylim(0.86, 1.01)
plt.ylabel('Mean accuracy')
plt.xlabel('Max tree depth')
plt.title('4.1(c) Gini acurracy')

plt.show()
```
